```python
from sklearn.metrics import classification_report

def predict(train,w=0.5):

  lst_real = []
  lst_pred = []

  for row in train:
      obs = {'B' : row[1], 'C' : row[2], 'D' : row[3]}
      pa = model.predict_proba(obs)[0].parameters[0][1]
      realA  = row[0]
      predA = 0
      if pa > w:
          predA = 1

      lst_pred.append(predA)
      lst_real.append(realA)

  print(lst_real,lst_pred)
  print(classification_report(lst_real,lst_pred))
  return 

predict(train)
```

```python
from sklearn.metrics import confusion_matrix

def cm(train,w=0.5):

  lst_real = []
  lst_pred = []

  for row in train:
      obs = {'B' : row[1], 'C' : row[2], 'D' : row[3]}
      pa = model.predict_proba(obs)[0].parameters[0][1]
      realA  = row[0]
      predA = 0
      if pa > w:
          predA = 1

      lst_pred.append(predA)
      lst_real.append(realA)

  confm = confusion_matrix(lst_real,lst_pred)
  print(confm)

  return confm

# Compute confusion matrix
confm = cm(test)
print('Confusion matrix - Bayesian Natework')
#print(cm_test)

print()

#Test set
L = 59
ACC = (confm[0][0] + confm[1][1]) / L
PREC = (confm[0][0])/ (confm[0][0]+ confm[1][0])
TPR = (confm[0][0]) / (confm[0][0] + confm[0][1])
TNR = (confm[1][1])/ (confm[1][1] + confm[1][0])
FPR = 1-TNR
REC = (confm[0][0])/ (confm[0][0] + confm[0][1])
F1SCORE = 2 * (REC * PREC)/(REC+PREC)
BAL_ACC = (REC+TNR)/2

print('Accuracy for test set = ', ACC)
print('Precision for test = ', PREC)
print('True Positive Rate = ', TPR)
print('True Negative Rate = ', TNR)
print('False Positive Rate = ', FPR)
print('Recall = ', REC)
print('F1 Score = ', F1SCORE)
print('Balanced Accuracy = ',BAL_ACC)
```

# **2.1 Importing The Data Set and Customizing**

```python
import pandas as pd

heartDisease = pd.read_csv('heart_cleveland_upload.csv')

del heartDisease['ca']
del heartDisease['slope']
del heartDisease['thal']
del heartDisease['oldpeak']

heartDisease = heartDisease.replace('?', np.nan)
heartDisease.dtypes
heartDisease.head()
```

# **2.2 Modeling The Data Set**

```python
model = BayesianModel([('age', 'trestbps'), ('age', 'fbs'), ('sex', 'trestbps'), ('sex', 'trestbps'), 
                       ('exang', 'trestbps'),('trestbps','condition'),('fbs','condition'),
                      ('condition','restecg'),('condition','thalach'),('condition','chol')])

# Learing CPDs using Maximum Likelihood Estimators
model.fit(heartDisease, estimator=MaximumLikelihoodEstimator)

```

```python
print(model.get_cpds('age'))
```

```python
print(model.get_cpds('chol'))
```

```python
print(model.get_cpds('sex'))
```

```python
model.nodes()
```

```python
model.edges()
```

# **2.3 Inferencing with Bayesian Network**

```python
# Doing exact inference using Variable Elimination
from pgmpy.inference import VariableElimination
HeartDisease_infer = VariableElimination(model)

q = HeartDisease_infer.query(variables=['condition'], evidence={'age': 29})
print(q)
```
